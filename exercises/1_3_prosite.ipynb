## Accessing PROSITE using Biopython

Access PROSITE, a database of protein domains, families, and functional sites defined by patterns (regular expressions) and profiles (position-specific scoring matrices). Two complementary modules:

- **`Bio.ExPASy.Prosite`** — fetch and parse PROSITE records (pattern definitions, documentation).
- **`Bio.ExPASy.ScanProsite`** — scan a protein sequence against the PROSITE database to find motif/domain hits.

### Key Functions

| Function | Purpose |
|----------|---------|
| `ExPASy.get_prosite_raw(accession)` | Fetch a PROSITE entry (e.g., "PS00005") in raw format |
| `Prosite.read(handle)` | Parse a PROSITE record |
| `Prosite.parse(handle)` | Parse multiple PROSITE records (iterator) |
| `ExPASy.get_prodoc_entry(accession)` | Fetch PROSITE documentation in HTML |
| `ScanProsite.scan(seq=...)` | Scan a sequence against PROSITE online |
| `ScanProsite.read(handle)` | Parse ScanProsite XML results |


## Minimal examples

**Scan a protein sequence against PROSITE:**

```python
from Bio.ExPASy import ScanProsite
import io

# Scan human insulin precursor sequence (UniProt P01308) against PROSITE
sequence = ("MALWMRLLPLLALLALWGPDPAAAFVNQHLCGSHLVEALYLVCGERGFFYTPKT"
            "RREAEDLQVGQVELGGGPGAGSLQPLALEGSLQKRGIVEQCCTSICSLYQLENYCN")

handle = ScanProsite.scan(seq=sequence)
raw = handle.read()

# If the server omits the XML declaration, prepend it
if not raw.startswith(b"<?xml"):
    raw = b'<?xml version="1.0" encoding="UTF-8"?>\n' + raw

result = ScanProsite.read(io.BytesIO(raw))

print(f"Number of sequences scanned: {result.n_seq}")
print(f"Number of matches found: {result.n_match}")

for hit in result:
    print(f"  PROSITE ID: {hit['signature_ac']}, "
          f"Position: {hit['start']}-{hit['stop']}")
```

```python
from Bio import ExPASy
from Bio.ExPASy import Prosite

handle = ExPASy.get_prosite_raw("PS00008")  # N-myristoylation site
record = Prosite.read(handle)

print(f"Accession: {record.accession}")
print(f"Name: {record.name}")
print(f"Description: {record.description}")
print(f"Pattern: {record.pattern}")
print(f"Type: {record.type}")
```
