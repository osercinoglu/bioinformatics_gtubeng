**Exercise W1: ExPASy ProtParam** (https://web.expasy.org/protparam/)

- Paste the bacteriorhodopsin sequence (Uniprot ID: **P02945**)
- Note the MW, theoretical pI, and GRAVY index (positive = overall hydrophobic)
- Compare with hemoglobin α-chain GRAVY (should be negative = overall hydrophilic)
- **What does GRAVY tell you about where each protein might be located in the cell?"**

**Exercise W2: ExPASy ProtScale** (https://web.expasy.org/protscale/)

- Select Kyte-Doolittle scale, window size = 19
- Paste bacteriorhodopsin → identify the 7 peaks above the 1.6 threshold line
- Change to window = 7 → observe how the plot becomes noisy
- Paste hemoglobin α-chain with window = 19 → no peaks above 1.6
