#### Setup: Download sequences

```bash
# Create a working directory
mkdir -p ~/protein_prediction && cd ~/protein_prediction

# Download sequences from UniProt (one-liners)
curl -s "https://rest.uniprot.org/uniprotkb/P02945.fasta" > bacteriorhodopsin.fasta
curl -s "https://rest.uniprot.org/uniprotkb/P69905.fasta" > hemoglobin_alpha.fasta
curl -s "https://rest.uniprot.org/uniprotkb/P01308.fasta" > insulin.fasta
curl -s "https://rest.uniprot.org/uniprotkb/P04637.fasta" > p53.fasta

# Quick check
head -2 bacteriorhodopsin.fasta
```

#### Exercise C1: Basic protein statistics with EMBOSS `pepstats`

```bash
# Run pepstats on bacteriorhodopsin
pepstats -sequence bacteriorhodopsin.fasta -outfile bact_stats.txt

# View the output
cat bact_stats.txt

# 1. What is the molecular weight?
# 2. What percentage of residues are "Tiny" vs "Small" vs "Large"?
# 3. What is the isoelectric point?
```

```bash
# Compare with hemoglobin alpha
pepstats -sequence hemoglobin_alpha.fasta -outfile hemo_stats.txt

# Side-by-side comparison of key properties
echo "=== Bacteriorhodopsin ===" && grep "Molecular weight" bact_stats.txt
echo "=== Hemoglobin alpha ===" && grep "Molecular weight" hemo_stats.txt
```

#### Exercise C2: Hydropathy plot with EMBOSS `pepwindow`

```bash
# Kyte-Doolittle hydropathy plot — default window = 19
pepwindow -sequence bacteriorhodopsin.fasta -graph svg -goutfile bact_hydro

# Try different window sizes
pepwindow -sequence bacteriorhodopsin.fasta -length 7 -graph svg -goutfile bact_hydro_w7
pepwindow -sequence bacteriorhodopsin.fasta -length 21 -graph svg -goutfile bact_hydro_w21

# Compare soluble vs membrane protein
pepwindow -sequence hemoglobin_alpha.fasta -graph svg -goutfile hemo_hydro
```

**Quick check:**

1. Open the SVG files (right-click in File Browser and select "Open in Browser Tab") and compare the w=7 vs w=19 vs w=21 plots for bacteriorhodopsin
2. Count the peaks above the ~1.6 threshold in the w=19 plot — how many TM helices?
3. Look at the hemoglobin plot — does it have any peaks above 1.6?

#### Exercise C3: Multi-property view with EMBOSS `pepinfo`

```bash
# pepinfo gives hydrophobicity, charge, and structural properties in one view
pepinfo -sequence bacteriorhodopsin.fasta -graph svg -goutfile bact_pepinfo
pepinfo -sequence p53.fasta -graph svg -goutfile p53_pepinfo
```

**Quick check:**

1. In the p53 pepinfo plot, identify regions that are mostly "Polar" / "Charged" — these may correspond to intrinsically disordered regions
2. Compare with the bacteriorhodopsin plot — is there a clear difference in the property distributions?
