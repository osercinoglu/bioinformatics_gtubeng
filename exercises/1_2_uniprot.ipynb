## Biopython's UniProt module

Biopython includes a module (Bio.UniProt) to access UniProt/SwissProt database programmatically.

For more information, visit the Biopython Uniprot documentation/tutorial at https://biopython.org/docs/latest/Tutorial/chapter_uniprot.html

Access UniProtKB (Swiss-Prot + TrEMBL) for protein sequences, annotations, functional data, cross-references, and more. Two approaches available in Biopython:

- **New (recommended):** `Bio.UniProt.search()` — wraps UniProt's REST API, returns tabular results using UniProt query syntax.
- **Classic:** `Bio.ExPASy.get_sprot_raw()` — fetches individual Swiss-Prot flat-file records by accession.

### Key Functions

| Function / Module | Purpose |
|-------------------|---------|
| `UniProt.search(query)` | Search UniProtKB, returns iterator of results (Biopython ≥1.84) |
| `ExPASy.get_sprot_raw(id)` | Fetch a single Swiss-Prot record in flat-file format |
| `SwissProt.read(handle)` | Parse a Swiss-Prot flat-file record into a Record object |
| `SeqIO.read(handle, "swiss")` | Parse Swiss-Prot record into a SeqRecord |
| `SeqIO.read(handle, "uniprot-xml")` | Parse UniProt XML into a SeqRecord |

## Minimal examples

Let's start with the search function:

```python
from Bio import UniProt

# Search for reviewed human insulin entries
results = list(UniProt.search("(gene:INS) AND (organism_id:9606) AND (reviewed:true)"))
```

```python
for entry in results[:2]:
    print(entry)
```

The classical approach:

```python
from Bio import ExPASy, SwissProt

# Fetch human p53 (UniProt accession P04637)
handle = ExPASy.get_sprot_raw("P04637")
record = SwissProt.read(handle)

print(f"Entry name: {record.entry_name}")
print(f"Organism: {record.organism}")
print(f"Sequence length: {record.sequence_length}")
print(f"Description: {record.description}")
print(f"Number of cross-references: {len(record.cross_references)}")
print(f"Keywords: {record.keywords}")
```

You can thus easily integrate UniProt-based database search & retrieval into your custom analysis workflow from within a Python script.

## REST API alternative

```python
import requests

# Fetch a protein entry in JSON format
url = "https://rest.uniprot.org/uniprotkb/P04637"
response = requests.get(url, headers={"Accept": "application/json"})
data = response.json()
print(f"Protein: {data['proteinDescription']['recommendedName']['fullName']['value']}")
print(f"Gene: {data['genes'][0]['geneName']['value']}")
```
